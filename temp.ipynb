```python
import requests
import pandas as pd
```

```python
payload = {
        'applicationId':'1035981462778848427',
        'genreId':200162
        }
url = 'https://app.rakuten.co.jp/services/api/IchibaItem/Ranking/20170628'
```

```python
r = requests.get(url, params=payload)
```

```python
resp = r.json()
```

```python
resp
```

```python
count = 0
for i in resp['Items']:
    count += 1
    print(f"[{count}位]")
    print("商品名 : " + i['Item']['itemName'])
    print("商品コード　：　" + str(i['Item']['itemCode'])) 
    print("値段　：　" + str(i['Item']['itemPrice']) + "円")
    print("URL　：　" + str(i['Item']['itemUrl'])) 
    print("\n")
```
